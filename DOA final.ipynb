```python
from pyspark.sql import SparkSession
from pyspark.ml.stat import Correlation
from pyspark.ml.feature import VectorAssembler
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Start Spark session
spark = SparkSession.builder.appName("CorrelationMatrix").getOrCreate()

# Define your numeric columns for correlation
numeric_columns = [
    'passenger_count',
    'trip_distance',
    'fare_amount',
    'extra',
    'mta_tax',
    'tip_amount',
    'tolls_amount',
    'total_amount',
    'improvement_surcharge',
    'trip_duration',
    'hour',
    'day'
]

# Assemble the columns into a feature vector
assembler = VectorAssembler(inputCols=numeric_columns, outputCol="features")
df_vector = assembler.transform(uber_df_cleaned).select("features")

# Calculate the correlation matrix
correlation_matrix = Correlation.corr(df_vector, "features").head()[0]

# Convert to a Pandas DataFrame for visualization
corr_array = correlation_matrix.toArray()
corr_df = pd.DataFrame(corr_array, index=numeric_columns, columns=numeric_columns)
```

```python
# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_df, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Numeric Features")
plt.tight_layout()
plt.show()
```

```python
from pyspark.ml.regression import LinearRegression, DecisionTreeRegressor, RandomForestRegressor
from pyspark.ml.regression import GBTRegressor

data = uber_df_cleaned.select("trip_duration","tip_amount","tolls_amount", "total_amount", "fare_amount", "trip_distance", "passenger_count", "hour", "day", "speed")

# Split the data into training and testing sets
train_data, test_data = data.randomSplit([0.8, 0.2], seed=42)

quantiles = test_data.approxQuantile("fare_amount", [0.25, 0.75], 0.01)
Q1 = quantiles[0]
Q3 = quantiles[1]
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

train_data_filtered = train_data.filter((F.col("fare_amount") >= lower_bound) & (F.col("fare_amount") <= upper_bound))
test_data_filtered = test_data.filter((F.col("fare_amount") >= lower_bound) & (F.col("fare_amount") <= upper_bound))
print(f"Original train data count: {train_data.count()}")
print(f"Filtered train data count: {train_data_filtered.count()}")
print(f"Original test data count: {test_data.count()}")
print(f"Filtered test data count: {test_data_filtered.count()}")
# Create a VectorAssembler to combine feature columns into a single vector
feature_columns = ["trip_distance", "passenger_count", "hour", "day", "speed"]
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
training_data = assembler.transform(test_data_filtered)
```

```python
from pyspark.ml import Pipeline
models = {
    "Decision Tree Regression": DecisionTreeRegressor(labelCol="fare_amount", featuresCol="features"),
    "Random Forest Regression": RandomForestRegressor(labelCol="fare_amount", featuresCol="features"),
    "Gradient Boosting Regression": GBTRegressor(labelCol="fare_amount", featuresCol="features")
}

# Create a dictionary to hold the trained models and their corresponding metrics
model_results = {}
predictions_list = []

# Loop through each model, train it, and evaluate
for model_name, model in models.items():
    pipeline = Pipeline(stages=[assembler, model])  # Assembler is assumed to be defined elsewhere

    # Fit the model
    model_trained = pipeline.fit(train_data_filtered)

    # Make predictions
    predictions = model_trained.transform(test_data_filtered)

    # Evaluate the model (RMSE and R²)
    evaluator = RegressionEvaluator(labelCol="fare_amount", predictionCol="prediction", metricName="rmse")
    rmse = evaluator.evaluate(predictions)
    evaluator_r2 = RegressionEvaluator(labelCol="fare_amount", predictionCol="prediction", metricName="r2")
    r2 = evaluator_r2.evaluate(predictions)

    # Store the results
    model_results[model_name] = {"RMSE": rmse, "R²": r2}

    # Store predictions for comparison later
    predictions_pd = predictions.select("fare_amount", "prediction").toPandas()
    predictions_pd["Model"] = model_name
    predictions_list.append(predictions_pd)

# Concatenate predictions from all models
all_predictions = pd.concat(predictions_list, ignore_index=True)

# Display model results
print("Model Evaluation Results:")
for model_name, metrics in model_results.items():
    print(f"{model_name}: RMSE = {metrics['RMSE']:.2f}, R² = {metrics['R²']:.2f}")

# Visualize performance metrics (RMSE and R²)
metrics_df = pd.DataFrame(model_results).T
metrics_df.plot(kind="bar", figsize=(10, 6), title="Model Comparison (RMSE and R²)")
plt.xticks(rotation=45)
plt.ylabel("Metric Value")
plt.tight_layout()
plt.show()

# Plot Actual vs Predicted for each model
plt.figure(figsize=(12, 6))
for model_name in model_results.keys():
    model_predictions = all_predictions[all_predictions["Model"] == model_name]
    plt.scatter(model_predictions["fare_amount"], model_predictions["prediction"], label=model_name, alpha=0.6)

# Plot a reference line (y = x) for perfect prediction
plt.plot([all_predictions["fare_amount"].min(), all_predictions["fare_amount"].max()],
         [all_predictions["fare_amount"].min(), all_predictions["fare_amount"].max()],
         color="black", linestyle="--", label="Perfect Prediction")

# Customize the plot
plt.title("Actual vs Predicted Fare Amounts")
plt.xlabel("Actual Fare Amount")
plt.ylabel("Predicted Fare Amount")
plt.legend(title="Model")
plt.tight_layout()
plt.show()
```

```python
from pyspark.sql import functions as F
from pyspark.sql.types import IntegerType, FloatType, TimestampType
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import DecisionTreeClassifier, LinearSVC
from pyspark.ml.evaluation import BinaryClassificationEvaluator

# Step 1: Data Cleaning (based on previous steps)

# Ensure data types are set correctly for feature engineering
uber_df_cleaned = uber_df_cleaned.withColumn("tpep_pickup_datetime", F.col("tpep_pickup_datetime").cast(TimestampType())) \
       .withColumn("tpep_dropoff_datetime", F.col("tpep_dropoff_datetime").cast(TimestampType())) \
       .withColumn("trip_distance", F.col("trip_distance").cast(FloatType())) \
       .withColumn("fare_amount", F.col("fare_amount").cast(FloatType())) \
       .withColumn("tip_amount", F.col("tip_amount").cast(FloatType())) \
       .withColumn("total_amount", F.col("total_amount").cast(FloatType())) \
       .withColumn("passenger_count", F.col("passenger_count").cast(IntegerType()))

# Remove invalid data where critical fields are zero or null
uber_df_cleaned = uber_df_cleaned.filter((F.col("trip_distance") > 0) &
               (F.col("fare_amount") > 0) &
               (F.col("passenger_count") > 0) &
               F.col("total_amount").isNotNull() &
               F.col("tpep_pickup_datetime").isNotNull() &
               F.col("tpep_dropoff_datetime").isNotNull())

# Step 2: Feature Engineering

# Calculate trip duration in minutes
uber_df_cleaned = uber_df_cleaned.withColumn("trip_duration", 
                   (F.unix_timestamp("tpep_dropoff_datetime") - F.unix_timestamp("tpep_pickup_datetime")) / 60)

# Calculate additional features
uber_df_cleaned = uber_df_cleaned.withColumn("fare_per_mile", F.col("fare_amount") / F.col("trip_distance")) \
       .withColumn("fare_per_minute", F.col("fare_amount") / F.col("trip_duration")) \
       .withColumn("tip_percentage", (F.col("tip_amount") / F.col("fare_amount")) * 100)

# Define a binary label for high-earning trips
high_earning_threshold = 50  # Threshold can be adjusted
uber_df_cleaned = uber_df_cleaned.withColumn("high_earning", F.when(F.col("total_amount") > high_earning_threshold, 1).otherwise(0))

# Step 3: Fill Null Values for Feature Columns

# Define columns to fill with 0 where nulls are present
feature_cols = ["trip_distance", "trip_duration", "fare_per_mile", "fare_per_minute", "tip_percentage"]

# Fill nulls in specified columns with 0
uber_df_cleaned = uber_df_cleaned.fillna(0, subset=feature_cols)

# Step 4: Vector Assembly for MLlib

assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
data = assembler.transform(uber_df_cleaned)

# Select only the features and label for modeling
data = data.select("features", "high_earning")

# Split the data into training and test sets
train_data, test_data = data.randomSplit([0.7, 0.3], seed=42)

# Step 5: Decision Tree Model Training and Evaluation

# Initialize Decision Tree model
dt = DecisionTreeClassifier(labelCol="high_earning", featuresCol="features")

# Train the model
dt_model = dt.fit(train_data)

# Make predictions
dt_predictions = dt_model.transform(test_data)

# Evaluate the model
evaluator = BinaryClassificationEvaluator(labelCol="high_earning", metricName="areaUnderROC")
dt_auc = evaluator.evaluate(dt_predictions)
print(f"Decision Tree AUC: {dt_auc}")

# Step 6: SVM Model Training and Evaluation

# Initialize Linear SVM model (PySpark only supports LinearSVC)
svm = LinearSVC(labelCol="high_earning", featuresCol="features", maxIter=10)

# Train the model
svm_model = svm.fit(train_data)

# Make predictions
svm_predictions = svm_model.transform(test_data)

# Evaluate the model
svm_auc = evaluator.evaluate(svm_predictions)
print(f"SVM AUC: {svm_auc}")
```
